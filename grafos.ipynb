```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(['João', 'Paulo', 'Maria', 'Joana', 'Antonia', 'Lili', 'Raimundo'])

```

```python
G.add_edges_from([
    ('João', 'Paulo'),
    ('Paulo', 'Maria'),
    ('Paulo', 'Joana'),
    ('Paulo', 'Antonia'),
    ('Paulo', 'Lili'),
    ('Maria', 'Joana'),
    ('Joana', 'Antonia'),
    ('Antonia', 'Lili'),])
```

```python
print(f"Nós do grafo: {G.nodes()}")
print(f"Arestas do grafro: {G.edges()}")
print(f"Contando o num de Nós: {G.number_of_nodes()}")
print(f"Contando o num de Arestas: {G.number_of_edges()}")
```

```python
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels= True, node_color='lightblue',
        font_weight='bold',node_size=700, font_size=10)

plt.title('Representação em Grafo simples')
plt.show()
```
