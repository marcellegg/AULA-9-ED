```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_nodes_from(['Esnupi', 'Fã-1', 'Fã-2', 'Fã-3', 'Não-Fã'])

```

```python
G.add_edges_from([
    ('Fã-1', 'Esnupi'),
    ('Fã-2', 'Esnupi'),
    ('Fã-3', 'Esnupi'),])

```

```python
print(f"Nós do grafo: {G.nodes()}")
print(f"Arestas do grafro: {G.edges()}")
print(f"Contando o num de Nós: {G.number_of_nodes()}")
print(f"Contando o num de Arestas: {G.number_of_edges()}")
```

```python
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels= True, node_color='lightgreen',
        node_size=2000, font_color='black', font_size=7, arrowsize=15)

plt.title('Representação em Grafo Direcionado')
plt.show()
```
